```python
f=open('StudentsPerformance.csv','r')
data=f.readlines()
student=[i.strip().split(',')for i in data]
print(student)        
```

```python

def group_avg_scores(data, group_col, score_col):
    header = data[0]
    group_idx = header.index(group_col)#"race/ethnicity"group_idx1
    score_idx = header.index(score_col)#"math score"score_idx5

    result = {}  # {'group A': [70, 80, 65], ...}

    for row in data[1:]:#header
        group = row[group_idx].strip('"')#groupheader index1
        score = int(row[score_idx].strip('"'))

        if group not in result:#group score
            result[group] = []
        result[group].append(score)

    #
    avg_result = {}
    for group, scores in result.items():
        avg = round(sum(scores) / len(scores), 2)
        avg_result[group] = avg

    return avg_result
```

```python
math_by_race = group_avg_scores(student, '"race/ethnicity"', '"math score"')
reading_by_race = group_avg_scores(student, '"race/ethnicity"', '"reading score"')
writing_by_race = group_avg_scores(student, '"race/ethnicity"', '"writing score"')

print("", math_by_race)
print("", reading_by_race)
print("", writing_by_race)
```

```python
math_by_edu = group_avg_scores(student, '"parental level of education"', '"math score"')
reading_by_edu = group_avg_scores(student, '"parental level of education"', '"reading score"')
writing_by_edu = group_avg_scores(student, '"parental level of education"', '"writing score"')

print("", math_by_edu)
print("", reading_by_edu)
print("", writing_by_edu)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

#  DataFrame
df_avg_by_race = pd.DataFrame({
    "math": math_by_race,
    "reading": reading_by_race,
    "writing": writing_by_race
}).reset_index().rename(columns={"index": "race/ethnicity"})

#  DataFrame
df_avg_by_edu = pd.DataFrame({
    "math": math_by_edu,
    "reading": reading_by_edu,
    "writing": writing_by_edu
}).reset_index().rename(columns={"index": "parental level of education"})

# 
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# 
df_avg_by_race.plot(kind="bar", x="race/ethnicity", ax=axes[0])
axes[0].set_title("Average Scores by Race/Ethnicity")
axes[0].set_ylabel("Score")
axes[0].set_xticklabels(df_avg_by_race["race/ethnicity"], rotation=30)

# 
df_avg_by_edu.plot(kind="bar", x="parental level of education", ax=axes[1])
axes[1].set_title("Average Scores by Parental Education")
axes[1].set_ylabel("Score")
axes[1].set_xticklabels(df_avg_by_edu["parental level of education"], rotation=30)

plt.tight_layout()
plt.savefig("score_comparison.png", dpi=300)  # 
plt.show()  #  jupyter cell 
```
